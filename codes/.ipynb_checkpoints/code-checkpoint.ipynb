## Removing unwanted letters, characters and numbers
Let's replace all unwanted letters with spaces.

```python
data["title"] = data["title"].str.replace("[^a-zA-Z]+", " ", case=False, regex=True)
data["text"] = data["text"].str.replace("[^a-zA-Z]+", " ", case=False, regex=True)
```

```python
data.head()
```

## Merge `title`, `text` and `subject` into a single column and remove all other rows except `true_news`

```python
data["title_text_subject"] = data['title'].apply(lambda x: str(x)) + " " +  data['text'].apply(lambda x: str(x)) + " " +  data['subject'].apply(lambda x: str(x))
```

```python
data.drop(columns = ["title","text","subject", "date"], inplace = True)
```

```python
data.head()
```

## Removing Stopwords

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
```

This function removes punctuations, before tokenizing our text and then lemmatizing the text after we removed stopwords.

```python
import string
import re

wn = nltk.WordNetLemmatizer()

def clean_text_lemmatize(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    # \W matches any non-word character (equivalent to [^a-zA-Z0-9_]). This does not include spaces i.e. \s
    # Add a + just in case there are 2 or more spaces between certain words
    clean_text = re.sub('[^a-zA-Z1-9]+', ' ', text)
    tokens = re.split('\W+', clean_text) # This function removes punctuations, before tokenizing our text and then stemming the text.
    
    # apply lemmatize and stopwords exclusion within the same step
    text = [wn.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

# Modelling

## Import Relevant Libraries

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score, plot_roc_curve, roc_auc_score, recall_score, precision_score, f1_score
```

## Set our X and y variable

```python
X = X_counts_df
y = data['true_news']
```

## Train Test Split our data to 80/20

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

## Create a Function to Run our Models

```python
def model_metrics(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    
    # R2 Scores
    print("Train R2 Score: " + str(accuracy_score(y_train, model.predict(X_train)))) # Train Score
    print("Test R2 Score: " + str(accuracy_score(y_test, model.predict(X_test)))) # Test Score
    
    # Cross Validation Score
    print("Cross Validation Score: " + str(cross_val_score(model, X_train, y_train, cv=5).mean()))
    
    # Confusion Matrix
    print(confusion_matrix(y_test, model.predict(X_test)))
    print(classification_report(y_test, model.predict(X_test)))
    
    # AUC ROC Graph
    plot_roc_curve(model, X_test, y_test)
    plt.plot([0, 1], [0, 1],
         label='baseline', linestyle='--')
    plt.legend();
    
    return None
```

```python
BernNB = BernoulliNB(binarize=True)
MultiNB = MultinomialNB()
GausNB = GaussianNB()
LogReg = LogisticRegression()
```

```python
model_metrics(X_train, y_train, X_test, y_test, BernNB)
```

```python
model_metrics(X_train, y_train, X_test, y_test, MultiNB)
```

```python
model_metrics(X_train, y_train, X_test, y_test, GausNB)
```

```python
model_metrics(X_train, y_train, X_test, y_test, LogReg)
```

### Find the ideal Hyperparameter for KNN

```python
knn_scores = {}
knn_scores_list = []

k_range = range(1,10)

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    knn_scores[k] = metrics.accuracy_score(y_test,y_pred)
    knn_scores_list.append(metrics.accuracy_score(y_test,y_pred))
    
plt.plot(k_range, knn_scores_list)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

Appears that n_neighbors = 2 yields the best result.

```python
knn = KNeighborsClassifier(n_neighbors=2)
    
model_metrics(X_train, y_train, X_test, y_test, knn)
```

# Future Improvement

### Fake news vs Real news
Currently, we are using Kaggle Real or Fake news dataset, which has it's data manually flagged. This means that our data could potentially have human error or imputed bias in our dataset. Thus, the agency that provided the data is important to maintain the integrity of the model.

### Mislabel of Propaganda News as Real News
Remember that ultimately, the agency is the ones who determine which data is real or fake. They may introduce propaganda by deliberately labelling news that otherwise is false or over-inflated as real news. Remember that history is written by victors and one should always read the provenance.

### Further Classification of different news category
Our data is currently merged together with the topic. We may be able to train individual models for each category to tailor to audience and search engine providers even better.

# Conclusion
Our model is able to attain above 95% threshold classification of real and fake news within the given input dataset.
